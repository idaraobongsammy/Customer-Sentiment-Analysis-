```python
import pandas as pd

# Assuming the unzipped file is named 'ratings_Electronics (1).csv' based on the file list in /content/
try:
    df = pd.read_csv('/content/ratings_Electronics (1).csv', header=None, names=['user_id', 'product_id', 'rating', 'timestamp'])
    print("DataFrame loaded successfully.")
    print("\nFirst 5 rows of the dataset:")
    display(df.head())
    print("\nInformation about the dataset:")
    display(df.info())
except FileNotFoundError:
    print("Error: The dataset file was not found in the expected location.")
    print("Please check the file name and path after unzipping.")
except Exception as e:
    print(f"An error occurred while loading the dataset: {e}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Categorize ratings into sentiment
def categorize_sentiment(rating):
    if rating >= 4:
        return 'Positive'
    elif rating == 3:
        return 'Neutral'
    else:
        return 'Negative'

df['sentiment'] = df['rating'].apply(categorize_sentiment)

# Analyze the distribution of sentiment
sentiment_counts = df['sentiment'].value_counts()
print("\nSentiment Distribution:")
display(sentiment_counts)

# Visualize the sentiment distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='viridis')
plt.title('Distribution of Customer Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
import zipfile
import os

zip_path = '/content/drive/MyDrive/Dataset /Amazon product reviews .zip'
extract_path = '/content/'

# Create the extraction directory if it doesn't exist
os.makedirs(extract_path, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(f"Dataset extracted to: {extract_path}")
```
